```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

data= np.loadtxt('body_0')
t=  data[0,]
x0= data[1,]
x1= data[2,]
x2= data[3,]
v0= data[4,]
v1= data[5,]
v2= data[6,]

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.set_title("3D_Curve")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

figure = ax.plot(x0, x1, x2, c='r')

plt.show()
```
